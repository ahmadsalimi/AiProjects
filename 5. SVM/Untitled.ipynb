```python
import numpy as np
import pandas as pd
from pathlib import Path
import PIL
import glob
from matplotlib import image
import matplotlib.pyplot as plt
```

```python
classes = Path("PersianLPR/").glob('*')
```

```python
X = []
y = []
```

```python
for class_path in classes:
    for image_path in class_path.glob("*"):
        image_file = image.imread(image_path)
        y.append('ุณ' if class_path.name == "S" else 'ุต' if class_path.name == "W" else class_path.name)
        X.append(image_file.reshape(image_file.size))
X = np.array(X)
y = np.array(y)
```

```python
columns = {f'pixel{i}': X[:, i] for i in range(y.size}
columns
```

```python
df = pd.DataFrame({'target': y, **{f'pixel{i}': X[:, i] for i in range(X.shape[1])}})
df.head()
```
